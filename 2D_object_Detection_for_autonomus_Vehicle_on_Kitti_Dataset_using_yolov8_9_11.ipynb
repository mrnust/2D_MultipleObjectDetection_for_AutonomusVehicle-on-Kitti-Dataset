```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
df_v8 = pd.read_csv('/content/drive/My Drive/Kitti dataset 500/completeyolov8/runs/detect/train/results.csv')
df_v9 = pd.read_csv('/content/drive/My Drive/Kitti dataset 500/completeyolov9/completeyolov9/runs/detect/train/results.csv')
df_v11 = pd.read_csv('/content/drive/My Drive/Kitti dataset 500/completeyolov11/runs/detect/train/results.csv')

# Clean up column names by stripping spaces
df_v8.columns = df_v8.columns.str.strip()
df_v9.columns = df_v9.columns.str.strip()
df_v11.columns = df_v11.columns.str.strip()

# Plot mAP@0.5 over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/mAP50(B)'], label='YOLOv8 mAP@0.5')
plt.plot(df_v9['epoch'], df_v9['metrics/mAP50(B)'], label='YOLOv9 mAP@0.5')
plt.plot(df_v11['epoch'], df_v11['metrics/mAP50(B)'], label='YOLOv11 mAP@0.5')
plt.xlabel('Epochs')
plt.ylabel('mAP@0.5')
plt.title('mAP@0.5 Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv11)')
plt.legend()
plt.show()

# Plot mAP@0.5:0.95 over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/mAP50-95(B)'], label='YOLOv8 mAP@0.5:0.95')
plt.plot(df_v9['epoch'], df_v9['metrics/mAP50-95(B)'], label='YOLOv9 mAP@0.5:0.95')
plt.plot(df_v11['epoch'], df_v11['metrics/mAP50-95(B)'], label='YOLOv11 mAP@0.5:0.95')
plt.xlabel('Epochs')
plt.ylabel('mAP@0.5:0.95')
plt.title('mAP@0.5:0.95 Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Precision over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/precision(B)'], label='YOLOv8 Precision')
plt.plot(df_v9['epoch'], df_v9['metrics/precision(B)'], label='YOLOv9 Precision')
plt.plot(df_v11['epoch'], df_v11['metrics/precision(B)'], label='YOLOv11 Precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.title('Precision Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Recall over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/recall(B)'], label='YOLOv8 Recall')
plt.plot(df_v9['epoch'], df_v9['metrics/recall(B)'], label='YOLOv9 Recall')
plt.plot(df_v11['epoch'], df_v11['metrics/recall(B)'], label='YOLOv11 Recall')
plt.xlabel('Epochs')
plt.ylabel('Recall')
plt.title('Recall Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Validation Box Loss over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['val/box_loss'], label='YOLOv8 Box Loss')
plt.plot(df_v9['epoch'], df_v9['val/box_loss'], label='YOLOv9 Box Loss')
plt.plot(df_v11['epoch'], df_v11['val/box_loss'], label='YOLOv11 Box Loss')
plt.xlabel('Epochs')
plt.ylabel('Validation Box Loss')
plt.title('Validation Box Loss Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv11)')
plt.legend()
plt.show()
```

```python
# Plot Classification Loss over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['train/cls_loss'], label='YOLOv8 Classification Loss')
plt.plot(df_v9['epoch'], df_v9['train/cls_loss'], label='YOLOv9 Classification Loss')
plt.plot(df_v11['epoch'], df_v11['train/cls_loss'], label='YOLOv11 Classification Loss')
plt.xlabel('Epochs')
plt.ylabel('Classification Loss')
plt.title('Classification Loss Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Distribution Focal Loss over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['train/dfl_loss'], label='YOLOv8 DFL Loss')
plt.plot(df_v9['epoch'], df_v9['train/dfl_loss'], label='YOLOv9 DFL Loss')
plt.plot(df_v11['epoch'], df_v11['train/dfl_loss'], label='YOLOv11 DFL Loss')
plt.xlabel('Epochs')
plt.ylabel('Distribution Focal Loss')
plt.title('Distribution Focal Loss Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv11)')
plt.legend()
plt.show()
```

# Analysis and Comparison

# Loss Metrics:

**Training Box Loss**: YOLOv9 has the lowest training box loss (0.7988), indicating it is performing better in minimizing the error in bounding box predictions compared to YOLOv8 and YOLOv11.

**Training Classification Loss**: YOLOv9 also has the lowest classification loss (0.49609), suggesting it is better at classifying objects correctly.

**Training Distribution Focal Loss**: YOLOv9 has a relatively low focal loss (0.88838) compared to the others.

# Metrics (Precision, Recall, mAP):

**Precision**: YOLOv9 leads significantly in precision (0.89327), indicating it has a high rate of correct positive predictions compared to the others.

**Recall**: YOLOv9 has the highest recall (0.80262), meaning it detects a larger proportion of actual positive cases.

**mAP@0.5**: YOLOv9 achieves the highest mAP@0.5 (0.8829), showcasing its overall effectiveness in detecting objects.

**mAP@0.5:0.95:** YOLOv9 again performs best (0.69347), suggesting it maintains good performance across various IoU thresholds.

# Validation Loss Metrics:

**Validation Box Loss**: YOLOv9 has the lowest validation box loss (0.72964), indicating it generalizes better to unseen data compared to the other models.

**Validation Classification Loss:** YOLOv9 also performs the best here (0.45217).

**Validation Distribution Focal Loss**: YOLOv9 has the lowest validation focal loss (0.88101).
