```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
df_v8 = pd.read_csv('/content/drive/My Drive/Kitti dataset 500/completeyolov8/runs/detect/train/results.csv')
df_v9 = pd.read_csv('/content/drive/My Drive/Kitti dataset 500/completeyolov9t/runs/detect/train/results.csv')
df_v10 = pd.read_csv('/content/drive/My Drive/Kitti dataset 500/completeyolov10/runs/detect/train/results.csv')
df_v11 = pd.read_csv('/content/drive/My Drive/Kitti dataset 500/completeyolov11/runs/detect/train/results.csv')

# Clean up column names by stripping spaces
df_v8.columns = df_v8.columns.str.strip()
df_v9.columns = df_v9.columns.str.strip()
df_v10.columns = df_v10.columns.str.strip()
df_v11.columns = df_v11.columns.str.strip()

# Plot mAP@0.5 over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/mAP50(B)'], label='YOLOv8 mAP@0.5')
plt.plot(df_v9['epoch'], df_v9['metrics/mAP50(B)'], label='YOLOv9 mAP@0.5')
plt.plot(df_v10['epoch'], df_v10['metrics/mAP50(B)'], label='YOLOv10 mAP@0.5')
plt.plot(df_v11['epoch'], df_v11['metrics/mAP50(B)'], label='YOLOv11 mAP@0.5')
plt.xlabel('Epochs')
plt.ylabel('mAP@0.5')
plt.title('mAP@0.5 Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv10 vs YOLOv11)')
plt.legend()
plt.show()

# Plot mAP@0.5:0.95 over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/mAP50-95(B)'], label='YOLOv8 mAP@0.5:0.95')
plt.plot(df_v9['epoch'], df_v9['metrics/mAP50-95(B)'], label='YOLOv9 mAP@0.5:0.95')
plt.plot(df_v10['epoch'], df_v10['metrics/mAP50-95(B)'], label='YOLOv10 mAP@0.5:0.95')
plt.plot(df_v11['epoch'], df_v11['metrics/mAP50-95(B)'], label='YOLOv11 mAP@0.5:0.95')
plt.xlabel('Epochs')
plt.ylabel('mAP@0.5:0.95')
plt.title('mAP@0.5:0.95 Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv10 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Precision over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/precision(B)'], label='YOLOv8 Precision')
plt.plot(df_v9['epoch'], df_v9['metrics/precision(B)'], label='YOLOv9 Precision')
plt.plot(df_v10['epoch'], df_v10['metrics/precision(B)'], label='YOLOv10 Precision')
plt.plot(df_v11['epoch'], df_v11['metrics/precision(B)'], label='YOLOv11 Precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.title('Precision Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv10 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Recall over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['metrics/recall(B)'], label='YOLOv8 Recall')
plt.plot(df_v9['epoch'], df_v9['metrics/recall(B)'], label='YOLOv9 Recall')
plt.plot(df_v10['epoch'], df_v10['metrics/recall(B)'], label='YOLOv10 Recall')
plt.plot(df_v11['epoch'], df_v11['metrics/recall(B)'], label='YOLOv11 Recall')
plt.xlabel('Epochs')
plt.ylabel('Recall')
plt.title('Recall Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv10 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Validation Box Loss over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['val/box_loss'], label='YOLOv8 Box Loss')
plt.plot(df_v9['epoch'], df_v9['val/box_loss'], label='YOLOv9 Box Loss')
plt.plot(df_v10['epoch'], df_v10['val/box_loss'], label='YOLOv10 Box Loss')
plt.plot(df_v11['epoch'], df_v11['val/box_loss'], label='YOLOv11 Box Loss')
plt.xlabel('Epochs')
plt.ylabel('Validation Box Loss')
plt.title('Validation Box Loss Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv10 vs YOLOv11)')
plt.legend()
plt.show()
```

```python
# Plot Classification Loss over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['train/cls_loss'], label='YOLOv8 Classification Loss')
plt.plot(df_v9['epoch'], df_v9['train/cls_loss'], label='YOLOv9 Classification Loss')
plt.plot(df_v10['epoch'], df_v10['train/cls_loss'], label='YOLOv10 Classification Loss')
plt.plot(df_v11['epoch'], df_v11['train/cls_loss'], label='YOLOv11 Classification Loss')
plt.xlabel('Epochs')
plt.ylabel('Classification Loss')
plt.title('Classification Loss Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv10 vs YOLOv11)')
plt.legend()
plt.show()

# Plot Distribution Focal Loss over all epochs
plt.figure(figsize=(10, 6))
plt.plot(df_v8['epoch'], df_v8['train/dfl_loss'], label='YOLOv8 DFL Loss')
plt.plot(df_v9['epoch'], df_v9['train/dfl_loss'], label='YOLOv9 DFL Loss')
plt.plot(df_v10['epoch'], df_v10['train/dfl_loss'], label='YOLOv10 DFL Loss')
plt.plot(df_v11['epoch'], df_v11['train/dfl_loss'], label='YOLOv11 DFL Loss')
plt.xlabel('Epochs')
plt.ylabel('Distribution Focal Loss')
plt.title('Distribution Focal Loss Comparison Over Epochs (YOLOv8 vs YOLOv9 vs YOLOv10 vs YOLOv11)')
plt.legend()
plt.show()
```

# **Analysis and Comparison**

# **Loss Metrics**:

**Training Box Loss**:

- YOLOv9t has the lowest training box loss (1.12672), indicating it performs best in minimizing bounding box prediction errors compared to YOLOv8, YOLOv10n, and YOLOv11n.

**Training Classification Loss:**

- YOLOv11n achieves the lowest classification loss (0.86734), suggesting it excels at accurately classifying objects.

**Training Distribution Focal Loss:**

- YOLOv9t again has the lowest focal loss (0.81584), indicating better handling of difficult-to-classify samples compared to the others.

# **Metrics (Precision, Recall, mAP):**

**Precision:**

- YOLOv9t leads with a precision of 0.97983, demonstrating the highest rate of correct positive predictions among the models.

**Recall:**

- YOLOv9t also achieves the highest recall (0.78458), meaning it detects a larger proportion of actual positives compared to the others.

**mAP @ 0.5:**

- YOLOv11n takes the lead here with a mAP@0.5 of 0.65181, showcasing strong overall effectiveness in object detection.

**mAP @ 0.5:0.95:**

- YOLOv9t performs best across various IoU thresholds with a mAP@0.5:0.95 of 0.6972, indicating strong detection consistency.

# **Validation Loss Metrics:**

**Validation Box Loss:**

- YOLOv9t shows the lowest validation box loss (0.47088), suggesting better generalization to unseen data.

**Validation Classification Loss:**

- YOLOv11n has the lowest validation classification loss (0.79289), indicating strong accuracy on validation data.

**Validation Distribution Focal Loss:**

- YOLOv9t maintains its performance with the lowest validation focal loss (0.70913), reinforcing its consistency with hard samples.


# **Conclusion**:

YOLOv9t emerges as the top performer overall, excelling in Precision, Recall, and Distribution Focal Loss. YOLOv11n stands out in classification accuracy and
mAP@0.5, but YOLOv10n generally trails in most metrics.
